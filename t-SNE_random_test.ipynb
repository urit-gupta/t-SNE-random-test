```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Generate random data
np.random.seed(322899)
n_samples = 1000
n_features = 50
random_data = np.random.rand(n_samples, n_features)
labels = [int(np.floor(np.random.random_sample() * 6)) for i in range(0, n_samples)]
for i in range(0, n_samples):
    random_data[i][n_features-1] = labels[i]
```

```python
# Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, max_iter=300, random_state=42)
reduced_data = tsne.fit_transform(random_data)
```

```python
# Visualize the results
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels)
plt.title("t-SNE on Random Data")
plt.show()
```

```python
# Standardize the data
scaler = StandardScaler()
data_standardized = scaler.fit_transform(random_data)
```

```python
# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_standardized)
```

```python
# Visualize the results
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=labels)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot on Random Data')
plt.show()
```
